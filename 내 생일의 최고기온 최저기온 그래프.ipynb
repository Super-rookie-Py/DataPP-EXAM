# 데이터 처리 프로그래밍
### 산업공학과 201613365 박건우

- 내 생일의 최고기온 최저기온 그래프

```python
# 내생일의 최대 기온 값
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
birth_max=[]
birth_min=[]
for row in data:
    month = row[0].split('-')[1]
    day = row[0].split('-')[2]
    if row[-1] != '' and row[-2] !='':
        if month == "06" and day=='14' :
            birth_max.append(float(row[-1]))
            birth_min.append(float(row[-2]))
```

```python
# 한글 깨짐 현상 해결하기.
import platform
from matplotlib import font_manager, rc 
import matplotlib
# '-' 부호가 제대로 표시되게 하는 설정 matplotlib.rcParams['axes.unicode_minus'] = False
# 운영 체제마다 한글이 보이게 하는 설정 # 윈도우
if platform.system() == 'Windows':
    path = "c:\Windows\Fonts\malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name() 
    rc('font', family=font_name)
```

```python
plt.figure(figsize=(10,4), dpi= 100)
plt.plot(birth_max, color = 'hotpink', label = '최고기온')
plt.plot(birth_min, color = 'skyblue', label = '최저기온')
plt.title("내 생일의 최고기온과 최저기온 그래프")
plt.legend()
plt.show()
```

## 기상청 자료개방 포털에서 찾은 데이터

### 공항기상관측(AMOS)에서 얻은 자료
### 2010/06/24 ~ 2020/06/24의 10년간의 일별자료

- 지점, 지점명, 일시, 평균풍속(KT), 최고기온, 최저기온, 강수량합에 대한 정보

```python
# 제주도에서 내생일의 최대 기온 값
import csv
import matplotlib.pyplot as plt

f = open('Jeju.csv')
data = csv.reader(f)
head = next(data)
# 칼럼 이름 확인
print(head)
```

```python
birth_max=[]
birth_min=[]
for row in data:
    month = row[2].split('-')[1]
    day = row[2].split('-')[2]
    if row[-3] != '' and row[-2] !='':
        if month == "06" and day=='14' :
            birth_max.append(float(row[-3]))
            birth_min.append(float(row[-2]))
```

```python
plt.figure(figsize=(10,4), dpi= 100)
plt.plot(birth_max, color = 'hotpink', label = '최고기온')
plt.plot(birth_min, color = 'skyblue', label = '최저기온')
plt.title("제주공항에서 내 생일의 최고기온과 최저기온 그래프")
plt.legend()
plt.show()
```
