# 데이터 처리 프로그래밍
### 산업공학과 201613365 박건우

- 서울에서 1월과 8월의 최고기온 히스토그램

```python
# 1월과 8월의 최대 기온 값

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug=[]
jan=[]
for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == "08":
            aug.append(float(row[-1]))
        elif month == "01":
            jan.append(float(row[-1]))
```

```python
plt.figure(figsize=(10,4), dpi= 100)
plt.hist(aug, bins=100, color="skyblue",label = '8월최고기온')
plt.hist(jan,bins=100, color = 'hotpink', label = '1월최고기온')
plt.title("1월과 8월의 최고기온 히스토그램")
plt.legend()
plt.show()
```

## 기상청 자료개방 포털에서 찾은 데이터

### 공항기상관측(AMOS)에서 얻은 자료
### 2010/06/24 ~ 2020/06/24의 10년간의 일별자료

- 지점, 지점명, 일시, 평균풍속(KT), 최고기온, 최저기온, 강수량합에 대한 정보

```python
# 제주도에서 1월과 8월의 최대 기온 값

import csv
import matplotlib.pyplot as plt

f = open('Jeju.csv')
data = csv.reader(f)
head = next(data)
# 칼럼 이름 확인
print(head)
```

```python
aug=[]
jan=[]
for row in data:
    month = row[2].split('-')[1]
    if row[-3] != '':
        if month == "08":
            aug.append(float(row[-3]))
        elif month == "01":
            jan.append(float(row[-3]))
```

```python
plt.figure(figsize=(10,4), dpi= 100)
plt.hist(aug, bins=100, color="g",label = '8월최고기온')
plt.hist(jan,bins=100, color = 'y', label = '1월최고기온')
plt.title("제주 공항에서 1월과 8월의 최고기온 히스토그램")
plt.legend()
plt.show()
```
